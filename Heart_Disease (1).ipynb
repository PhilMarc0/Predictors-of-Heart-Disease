# Defining the problem 
## Introduction 
Heart disease, encompassing a range of conditions such as coronary artery disease, heart attacks, heart failure, and arrhythmias, is a leading cause of death worldwide. Given its prevalence and impact, understanding and addressing the triggers and risk factors associated with heart disease is of paramount importance.

Heart disease is largely influenced by a combination of lifestyle, genetic, and environmental factors. Common triggers include high blood pressure, high cholesterol, smoking, diabetes, obesity, physical inactivity, and poor diet. Stress and excessive alcohol consumption also play significant roles. These factors can damage the heart and blood vessels, leading to plaque buildup, reduced blood flow, and ultimately, heart-related complications.

## Why should we care?
Knowing about the triggers of heart disease can 

- act as a form of **preventive measure** by identifying risk factors and pursuing effective intervention actions to reduce the likelihood of an occurrence of heart disease.
- facilitate **early detection efforts** by healthcare professionals to treat and manage the health outcomes of patients effectively
- inculcate a sense of accountability to improve one's longevity and quality of life to prevent the occurrence of heart diseases which could pose a **financial burden** on the one being diagnosed and even their family burdens, if regular screening and medications are required. 

## About the dataset

The dataset contains the features:

1. age: age of the patient
2. sex: Gender of the patient (0: female, 1: Male)
3. cp: Type of chest pain
4. trestbps: Resting Blood pressure
5. chol: Serum cholesterol
6. fbs: Fasting blood sugar > 120 mg/dl
7. restecg: Resting electrocardiographic results
8. thalach: Maximum heart rate achieved
9. exang: Exercise induced angina
10. oldpeak: ST depression induced by exercise relative to rest
11. slope: Slope of the peak exercise ST segment
    * 0: upsloping
    * 1: Flat
    * 2: Downsloping
12. ca: Number of major vessels (0-4) colored by fluoroscopy
13. thal: Thalium stress test result    * 0: Normal
    * 1: Fixe  defect
    * 
2: Reversible dfett
    * 
3: Not described

```python
#Subsetting the data based on the tresbps, chol, and thalach columns
df_partial = df[['age','chol','thalach']]
df_partial.head()
```

```python
df_partial.corr()
```

```python
X = df_partial
y = df['target']

#Adding a constant to independent variable (for the intercept)
X = sm.add_constant(X)

#Fit the logistic regression model
model = sm.Logit(y,X)
result = model.fit()
```

```python
print(result.summary())
```

As can be seen from the results, the maximum heart rate achieved is statistically significant at all levels of significance. Therefore, the maximum heart rate achieved is a critical factor in predicting if someone has heart disease. 

```python
#Evaluating model performance on a test set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 420)
model = sm.Logit(y_train, X_train).fit()
y_pred_prob = model.predict(X_test)
y_pred = [1 if x>0.5 else 0 for x in y_pred_prob]
#print classification report
print(classification_report(y_test, y_pred))
```

**Explanation of each evaluation metric**

Precision: Measures the number of positive identifications that were correctly predicted out of the positive class observations. <br>
Recall: Measures the number of observations correctly predicted out of all the correctly labeled observations. <br>
F1-score: A combination of precision and recall. A perfect model has an F1-score of 1.0 <br>
Support: Displays the number of samples each metric is based on. 

### Analysis of model performance

This model has a higher precision on class 0 (No heart disease) than class 1 (Has heart disease). This model performed better in correctly predicting patients with heart disease, in terms of recall, which is good in healthcare applications where precisely detecting patients with symptoms of heart disease, could provide insights to doctors on prescriptions to reduce the number of lives lost.
